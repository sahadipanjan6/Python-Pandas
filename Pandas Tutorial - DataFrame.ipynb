```python
# importing the required packages
import numpy as np
import pandas as pd
```

```python
# creating the following DataFrame using Pandas library

#                Population        GDP        Surface        HDI        Continent
# -------------------------------------------------------------------------------------------
#   Canada       35.467         1785387       9984670       0.913        America
#   France       63.951         2833687        640679       0.888        Europe
#   Germany      80.94          3874437        357114       0.916        Europe
#   Italy        60.665         2167744        301336       0.873        Europe
#   Japan       127.061         4602367        377930       0.891        Asia
#   U.K.         64.511         2950039        242495       0.907        Europe
#   U.S.A       318.523         17348075      9525067       0.915        America

# -------------------------------------------------------------------------------------------

my_dataframe = pd.DataFrame({
    
    'Population' : [35.467,
                    63.951,
                    80.94,
                    60.665,
                    127.061,
                    64.511,
                    318.523],
    
    'GDP' : [1785387,
             2833687,
             3874437,
             2167744,
             4602367,
             2950039,
            17348075],
    
    'Surface' : [9984670,
                 640679,
                 357114,
                 301336,
                 377930,
                 242495,
                 9525067],
    
    'HDI' : [0.913,
             0.888,
             0.916,
             0.873,
             0.891,
             0.907,
             0.915],
    
    'Continent' : ['America',
                   'Europe',
                   'Europe',
                   'Europe',
                   'Asia',
                   'Europe',
                   'America']
    
}
)

# displaying the created DataFrame
print("Original Dataframe created is as follows:- \n")
my_dataframe
```

```python
# setting the names of the indexes of the dataframe
my_dataframe.index = ['Canada',
                      'France',
                      'Germany',
                      'Italy',
                      'Japan',
                      'United Kingdom',
                      'United States']

# displaying the named DataFrame
print("\n The named Dataframe is as shown below:- \n")
my_dataframe
```

```python
# determining the various attributes of the dataframe

# 1. Columns of the dataframe
print("Columns of the dataframe are:", my_dataframe.columns, "\n")

# 2. Indexes of the dataframe
print("Indexes of the dataframe are:", my_dataframe.index, "\n")

# 3. Information about the dataframe
print("Information about the dataframe is shown below:- \n")
print(my_dataframe.info, "\n")

# 4. Size of the dataframe
print("Size of the dataframe:", my_dataframe.size, "\n")

# 5. Shape of the dataframe
print("Shape of the dataframe:", my_dataframe.shape, "\n")

# 6. Description of the dataframe
print("Description of the dataframe is shown below:- \n")
print(my_dataframe.describe(), "\n")

# 7. Datatypes present in the dataframe
print("Datatypes of the dataframe are shown below:- \n")
print(my_dataframe.dtypes, "\n")

# 8. Count of the values of particular datatypes
print("Count of values of different datatypes are:- \n")
print(my_dataframe.dtypes.value_counts(), "\n")
```

```python
# Indexing, Selection and Slicing of the DataFrame

# extracting a particular row from the dataframe using the index name itself
index_name = input("Enter the name of the index: ")
print("Data of the particular index is as shown below:- \n")
print(my_dataframe.loc[index_name], "\n")

# extracting a particular row from the dataframe based on the index position
index_position = int(input("Enter the position of the index: "))
print("Data of the particular index is as shown below:- \n")
print(my_dataframe.iloc[index_position], "\n")
```

```python
# displaying the values of a particular column
column_name = input("Enter the name of the column: ")
print("Values of", column_name, "are as follows:- \n")
print(my_dataframe[column_name], "\n")

# displaying mutiple columns
number_of_columns = int(input("Enter number of columns: "))
list_of_columns = []
print("Enter the column names one by one... \n")
for i in range(number_of_columns):
    list_of_columns.append(input())
    
print("Values of the entered columns are as follows:- \n")
print(my_dataframe[list_of_columns], "\n")
```

```python
# 1. SLICING the dataframe based on index positions
starting_index = int(input("Enter the starting index: "))
ending_index = int(input("Enter the ending index: "))
print("Sub-Dataframe is as follows:- \n")
print(my_dataframe.iloc[starting_index : ending_index], "\n")

# 2. SLICING the dataframe based on index names
starting_index_name = input("Enter the starting index name: ")
ending_index_name = input("Enter the ending index name: ")
print("Sub-Dataframe is as follows:- \n")
print(my_dataframe.loc[starting_index_name : ending_index_name], "\n")

# 3. SLICING the dataframe based on selected rows and selected columns
start_index_name = input("Enter starting index name: ")
end_index_name = input("Enter ending index name: ")
column_list = []
number_of_columns = int(input("Enter no. of columns: "))
print("Enter the column names..\n")
for j in range(number_of_columns):
    column_list.append(input())
    
print("Sub-Dataframe is as follows:- \n")
print(my_dataframe.loc[start_index_name : end_index_name, column_list])
```
